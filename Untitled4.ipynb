Maintenant, ce que l'on peut faire, c'est réduire la dimensionnalité de chaque variable en les regroupant.

Vu que l'on a 3 différentes variables, on pourrait regrouper les réponses de chaque variable en 1 dimension. Pour cela, on peut utiliser la méthode la plus connu qui est l'Analyse en Composantes Principales (PCA). Cela va nous permettre de passer de plusieurs dimensions (ce qui est très dur à traiter et à clusteriser ensuite) à 1 seule dimension.

```python
"""
On fait le PCA pour réduire les dimensions pour A9, A10, A11
"""
scaler = StandardScaler()
A9_normalized = scaler.fit_transform(A9)
A10_normalized = scaler.fit_transform(A10)
A11_normalized = scaler.fit_transform(A11)

A9_pca = PCA(n_components=1)
A10_pca = PCA(n_components=1)
A11_pca = PCA(n_components=1)

A9_pca_result = A9_pca.fit_transform(A9_normalized)
A10_pca_result = A10_pca.fit_transform(A10_normalized)
A11_pca_result = A11_pca.fit_transform(A11_normalized)

A9_pca_df = pd.DataFrame(A9_pca_result, columns=["PCA1_value"])
A10_pca_df = pd.DataFrame(A10_pca_result, columns=["PCA2_value"])
A11_pca_df = pd.DataFrame(A11_pca_result, columns=["PCA3_value"])

A9_pca_df.head(5)
```

#### K-Means

Le K-means est une technique de classification ayant pour objectif de séparer nos données en groupes homogènes ayant des caractéristiques communes. La variable K représente le nombre de groupes dans les données. Il est rapide et efficace, surtout pour des données réduites en 3 dimensions, et produit des clusters de taille similaire.

```python
"""
On va calculer la silhouette score de notre K-Means pour un nombre de cluster entre 2 et 10
"""
silhouette_scores_kmeans = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    lab = kmeans.fit_predict(data_cluster.iloc[:, :3])
    score = silhouette_score(data_cluster.iloc[:, :3], lab)
    silhouette_scores_kmeans.append(score)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores_kmeans, marker='o', linestyle='--')
plt.title("Silhouette score en fonction du nombre de clusters")
plt.xlabel("Nombre de clusters")
plt.ylabel("Silhouette Score")
plt.xticks(range(2, 11))
plt.grid()
plt.show()
```

On constate que notre meilleur K-Means est lorsqu'il y a seulement 2 clusters. Bien qu'une silhouette score de 0.4 est plutôt bon, faire seulement 2 clusters n'est pas quelque chose à faire pour de l'analyse et classer des personnes. Il faudrait affiner nos résultats en divisant dans plus de clusters les individus, pour pouvoir faire une distinction plus fine des consommateurs.

Nous pouvons choisir entre 4 et 6 clusters, mais 4 est peut être aussi un peu trop léger. Nous prendrons donc 6 clusters.
Essayons de voir notre K-Means graphiquement.

```python
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans_labels = kmeans.fit_predict(data_cluster)
data_cluster['KMeans_Label'] = kmeans_labels
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_cluster.iloc[:, 0], data_cluster.iloc[:, 1], data_cluster.iloc[:, 2],
           c=data_cluster['KMeans_Label'], cmap='viridis', s=50)
ax.set_title("K-Means Clustering")
ax.set_xlabel("PCA1 (A9)")
ax.set_ylabel("PCA2 (A10)")
ax.set_zlabel("PCA3 (A11)")
plt.show()
```

On constate 6 clusters dans notre graphe.

- Le vert clair qui contient les utilisateurs plutôt "très positif". Comme on l'a pu le constater sur les dataframes, l'utilisateur 1 a répondu tout le temps `Tout à fait d'accord` , et lorsqu'on regarde son score PCA, il est à -6 environ. On constate donc que le cluster vert clair contient les gens à réponses "très positive".
- A l'inverse, le vert foncé à droite contient les utilisateurs à réponses "négative", car leurs valeurs se trouvent à l'extrême positif.
- Quant au violet, il doit sûrement contenir les personnes entre "neutre" et "positif"
- Le vert foncé contient les personnes entre "neutre" et "négatif"
- Le bleu foncé et le jaune contiennent les personnes "neutres"

Regardons maintenant les variances intra-groupes et inter-groupes :

```python
global_centroid = data_cluster.iloc[:, :3].mean(axis=0).values
intra_variance = 0
inter_variance = 0

for i in np.unique(kmeans_labels):
    cluster_points = data_cluster[data_cluster['KMeans_Label'] == i].iloc[:, :3].values
    cluster_centroid = cluster_points.mean(axis=0)

    """
    On va calcuser la variance intra cluster
    """
    intra_variance += np.sum((cluster_points - cluster_centroid) ** 2)

    """
    On va calculer la variance inter cluster
    """
    inter_variance += len(cluster_points) * np.sum((cluster_centroid - global_centroid) ** 2)


ratio = inter_variance / intra_variance
print(f"Variance intra groupe : {intra_variance}")
print(f"Variance inter groupe : {inter_variance}")
print(f"Ratio inter/intra : {ratio}")
```

Notre variance inter groupe représente la somme des distances au carré des centroïdes des clusters par rapport au centroïde global.
Notre variance intra groupe représente la somme des distances au carré des points à leurs centroïdes respectifs.

Globalement, nos clusters sont bien formés: notre variance inter groupe est supérieur à notre variance intra groupe. La ratio > 1 exprime qu'on a bien des clusters distincts, mais pas parfaitement séparés.

#### DBSCAN

Le DBSCAN (Density-Based Spatial Clustering of Applications with Noise) est un algorithme simple qui définit des clusters en utilisant l’estimation de la densité locale. Ses points forts sont qu'il est capable de trouver des clusters de formes variées, et qu'il ne nécessite pas de spécifier un nombre de clusters.

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_cluster)
data_cluster['DBSCAN_Label'] = dbscan_labels
score = silhouette_score(data_cluster.iloc[:, :3], dbscan_labels)
print(f"Silhouette Score DBSCAN : {score}")
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_cluster.iloc[:, 0], data_cluster.iloc[:, 1], data_cluster.iloc[:, 2],
           c=data_cluster['DBSCAN_Label'], cmap='viridis', s=50)
ax.set_title("DBSCAN Clustering")
ax.set_xlabel("PCA1 (A9)")
ax.set_ylabel("PCA2 (A10)")
ax.set_zlabel("PCA3 (A11)")
plt.show()
```

On constate que DBSCAN ne marche absolument pas. Son silhouette score est carrément négatif, et sa gestion de cluster est catastrophique. Nous ne l'utiliserons pas.

Il utilise des arbes de clustering (CF Trees) pour réduire la complexité en comprimant les données en une structure compacte avant d'effectuer le cluster final.

```python
"""
On va calculer la silhouette score de notre Birch pour un nombre de cluster entre 2 et 10
"""
silhouette_scores_birch = []
for n_clusters in range(2, 11):
    birch = Birch(n_clusters=n_clusters)
    lab = birch.fit_predict(data_cluster.iloc[:, :3])
    score = silhouette_score(data_cluster.iloc[:, :3], lab)
    silhouette_scores_birch.append(score)
```

#### Agglomerative Clustering

L'agglomerative clustering est une méthode hiérarchique où chaque point commence comme un cluster individuel. Les clusters sont ensuite fusionnés itérativement en fonction de leur similarité jusqu'à ce que le nombre définit de cluster soit atteint.

```python
"""
On va calculer la silhouette score de notre Agglomerative Clustering pour un nombre de cluster entre 2 et 10
"""
silhouette_scores_ac = []
for n_clusters in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=n_clusters, linkage="ward")
    lab = agglo.fit_predict(data_cluster.iloc[:, :3])
    score = silhouette_score(data_cluster.iloc[:, :3], lab)
    silhouette_scores_ac.append(score)
```

Pour cela, on peut utiliser la méthode la plus connu qui est l'Analyse en Composantes Principales (PCA). Cela va nous permettre de passer de plusieurs dimensions (ce qui est très dur à traiter et à clusteriser ensuite) à 1 seule dimension.

```python
"""
On fait le PCA pour réduire les dimensions pour A9, A10, A11
"""
scaler = StandardScaler()
A_normalized = scaler.fit_transform(A)
B_normalized = scaler.fit_transform(B)
C_normalized = scaler.fit_transform(C)

A_pca = PCA(n_components=1)
B_pca = PCA(n_components=1)
C_pca = PCA(n_components=1)

A_pca_result = A_pca.fit_transform(A_normalized)
B_pca_result = B_pca.fit_transform(B_normalized)
C_pca_result = C_pca.fit_transform(C_normalized)

A_pca_df = pd.DataFrame(A_pca_result, columns=["PCA1_value"])
B_pca_df = pd.DataFrame(B_pca_result, columns=["PCA2_value"])
C_pca_df = pd.DataFrame(C_pca_result, columns=["PCA3_value"])

A_pca_df.head(5)
```

Maintenant, nous avons 2 choix :

- Soit concaténer les 3 dataframes pour faire 1 seule dimension pour le cluster.
- Soit garder les 3 dataframes pour faire un cluster à 3 dimensions.

Garder les 3 dimensions est préférable car cela préserve la richesse des données et leurs relations. Cela nous permet aussi de faire
une représentation visuelle en 3D de nos clusters, ce qui permet une meilleure compréhension.

Nous devons maintenant faire des méthodes de clusterisation sur nos données et les comparer pour garder la meilleure
clusterisation effectuée. Nous allons tester 4 méthodes de clusterisation :

- K-Means
- DBSCAN
- BIRCH
- Agglomerative Clustering

Nous allons à chaque fois pour chaque méthode tester différentes valeurs de clusters et d'hyperparamètres pour avoir les
meilleurs résultats possibles.

```python
data_cluster2 = pd.concat([A_pca_df, B_pca_df, C_pca_df], axis=1)
data_cluster2.head(5)
```

#### K-Means

Le K-means est une technique de classification ayant pour objectif de séparer vos données en groupes homogènes ayant des caractéristiques communes. La variable K représente le nombre de groupes dans les données. Il est rapide et efficace, surtout pour des données réduites en 3 dimensions, et produit des clusters de taille similaire.

```python
"""
On va calculer la silhouette score de notre K-Means pour un nombre de cluster entre 2 et 10
"""
silhouette_scores_kmeans2 = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    lab = kmeans.fit_predict(data_cluster2.iloc[:, :3])
    score = silhouette_score(data_cluster2.iloc[:, :3], lab)
    silhouette_scores_kmeans2.append(score)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores_kmeans2, marker='o', linestyle='--')
plt.title("Silhouette score en fonction du nombre de clusters")
plt.xlabel("Nombre de clusters")
plt.ylabel("Silhouette Score")
plt.xticks(range(2, 11))
plt.grid()
plt.show()
```

On constate que notre meilleur K-Means est lorsqu'il y a seulement 2 clusters comme pour la partie précedente. Bien qu'une silhouette score de 0.4 est plutôt bon, faire seulement 2 clusters n'est pas quelque chose à faire pour de l'analyse et classer des personnes. Il faudrait affiner nos résultats en divisant dans plus de clusters les individus, pour pouvoir faire une distinction plus fine des consommateurs.

Nous pouvons choisir entre 4 et 6 clusters, mais 4 est peut être aussi un peu trop léger. Nous prendrons donc 6 clusters.
Essayons de voir notre K-Means graphiquement.

```python
kmeans2 = KMeans(n_clusters=6, random_state=42)
kmeans_labels2 = kmeans2.fit_predict(data_cluster2)
data_cluster2['KMeans_Label'] = kmeans_labels2
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_cluster2.iloc[:, 0], data_cluster2.iloc[:, 1], data_cluster2.iloc[:, 2],
           c=data_cluster2['KMeans_Label'], cmap='viridis', s=50)
ax.set_title("K-Means Clustering")
ax.set_xlabel("PCA1 (A)")
ax.set_ylabel("PCA2 (B)")
ax.set_zlabel("PCA3 (C)")
plt.show()
```

On constate 6 clusters assez regroupés dans notre graphe.

- Le vert qui contient les utilisateurs plutôt "très actif". Comme on a pu le constater sur les dataframes, lorsque l'utilisateur répond aux réponses de la variable B, la valeur `Jamais` correspond à 4 et la valeur `A chaque fois que vous jardinez ou presque` , correspond à 0.
- A l'inverse, le jaune et vert clair en haut et à droite contiennent les utilisateurs à réponses "peu actif", car leurs valeurs se trouvent à l'extrême positif.
- Quant au vert foncé, bleu foncé (que l'on voit presque pas) et violet, ils doivent sûrement contenir les personnes "moyen actif"

Regardons maintenant les variances intra-groupes et inter-groupes :

```python
global_centroid = data_cluster2.iloc[:, :3].mean(axis=0).values
intra_variance = 0
inter_variance = 0

for i in np.unique(kmeans_labels2):
    cluster_points = data_cluster2[data_cluster2['KMeans_Label'] == i].iloc[:, :3].values
    cluster_centroid = cluster_points.mean(axis=0)

    """
    On va calcuser la variance intra cluster
    """
    intra_variance += np.sum((cluster_points - cluster_centroid) ** 2)

    """
    On va calculer la variance inter cluster
    """
    inter_variance += len(cluster_points) * np.sum((cluster_centroid - global_centroid) ** 2)


ratio = inter_variance / intra_variance
print(f"Variance intra groupe : {intra_variance}")
print(f"Variance inter groupe : {inter_variance}")
print(f"Ratio inter/intra : {ratio}")
```

Notre variance inter groupe représente la somme des distances au carré des centroïdes des clusters par rapport au centroïde global.
Notre variance intra groupe représente la somme des distances au carré des points à leurs centroïdes respectifs.

Globalement, nos clusters sont bien formés: notre variance inter groupe est supérieur à notre variance intra groupe. La ratio > 1 exprime qu'on a bien des clusters distincts, mais pas parfaitement séparés.

#### DBSCAN

Le DBSCAN (Density-Based Spatial Clustering of Applications with Noise) est un algorithme simple qui définit des clusters en utilisant l’estimation de la densité locale. Ses points forts sont qu'il est capable de trouver des clusters de formes variées, et qu'il ne nécessite pas de spécifier un nombre de clusters.

```python
dbscan2 = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels2 = dbscan2.fit_predict(data_cluster2)
data_cluster2['DBSCAN_Label'] = dbscan_labels2
score2 = silhouette_score(data_cluster2.iloc[:, :3], dbscan_labels2)
print(f"Silhouette Score DBSCAN : {score2}")
```

#### Agglomerative Clustering

L'agglomerative clustering est une méthode hiérarchique où chaque point commence comme un cluster individuel. Les clusters sont ensuite fusionnés itérativement en fonction de leur similarité jusqu'à ce que le nombre définit de cluster soit atteint.

```python
"""
On va calculer la silhouette score de notre Agglomerative Clustering pour un nombre de cluster entre 2 et 10
"""
silhouette_scores_ac2 = []
for n_clusters in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=n_clusters, linkage="ward")
    lab = agglo.fit_predict(data_cluster2.iloc[:, :3])
    score = silhouette_score(data_cluster2.iloc[:, :3], lab)
    silhouette_scores_ac2.append(score)
```
